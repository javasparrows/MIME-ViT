```python
import matplotlib.pyplot as plt
import cv2
import json

# 
annotation_path = '/home/yukik/Work/Tohoku/Datasets/CMMD_0601_cleaned/D1-0033/R_MLO/AnnotationFile.Json'
image_path = '/home/yukik/Work/Tohoku/Datasets/CMMD_0601_cleaned/D1-0033/R_MLO/image.png'

# 
with open(annotation_path, 'r') as f:
    annotations = json.load(f)

# cgPoints
cg_points = annotations[0]['cgPoints']  # : annotations[0]

# 
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
img_rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)

# 
fig, ax = plt.subplots()
ax.imshow(img_rgb, cmap='gray')

# 
for point in cg_points:
    ax.plot(point['x'], point['y'], 'ro')  # 'ro' 

# 
plt.axis('off')  # 
plt.show()
```
